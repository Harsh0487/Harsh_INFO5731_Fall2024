# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
import math


def calculate_hemisphere_volume(radius):
    return (2/3) * math.pi * (radius ** 3)


even_radii = range(2, 21, 2)


for radius in even_radii:
    volume = calculate_hemisphere_volume(radius)
    print(f"Radius: {radius}, Volume: {volume:.2f}")

```

# Question 3 (20 points).

Create a Python program that encompasses various string manipulation task:

(a) Prompt the user to input two sentences.

(b) Transform each sentence into a list, where each element corresponds to a word in the sentence. For example, if the sentences are "Hello, how are you doing today?" and "This is an example sentence for the program!"

The program should output:
['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(c) Compare the two sentences. Print out a list of words that occur in both sentences.

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
import string

def get_sentences():
    """Prompt the user to input two sentences."""
    sentence1 = input("Please enter the first sentence: ")
    sentence2 = input("Please enter the second sentence: ")
    return sentence1, sentence2

def split_sentences_into_words(sentence1, sentence2):
    """Transform each sentence into a list, where each element corresponds to a word in the sentence."""
    words = sentence1.split() + sentence2.split()
    return words

def compare_sentences(sentence1, sentence2):
    """Compare the two sentences. Print out a list of words that occur in both sentences."""
    set1 = set(sentence1.split())
    set2 = set(sentence2.split())
    common_words = set1.intersection(set2)
    return list(common_words)

def remove_punctuation(sentence1, sentence2):
    """Remove any punctuation from the sentence and print the sanitized sentences."""
    translator = str.maketrans('', '', string.punctuation)
    sanitized_sentence1 = sentence1.translate(translator)
    sanitized_sentence2 = sentence2.translate(translator)
    return sanitized_sentence1, sanitized_sentence2

def count_vowels(sentence1, sentence2):
    """Count the number of vowels in each of the sentences."""
    vowels = 'aeiou'
    vowel_counts1 = {vowel: sentence1.lower().count(vowel) for vowel in vowels}
    vowel_counts2 = {vowel: sentence2.lower().count(vowel) for vowel in vowels}
    return vowel_counts1, vowel_counts2

def main():
    sentence1, sentence2 = get_sentences()
    print("\nYou entered the following sentences:")
    print("Sentence 1:", sentence1)
    print("Sentence 2:", sentence2)

    words = split_sentences_into_words(sentence1, sentence2)
    print("\nCombined list of words:")
    print(words)

    common_words = compare_sentences(sentence1, sentence2)
    print("\nCommon words:")
    print(common_words)

    sanitized_sentence1, sanitized_sentence2 = remove_punctuation(sentence1, sentence2)
    print("\nSanitized sentences:")
    print("Sentence 1:", sanitized_sentence1)
    print("Sentence 2:", sanitized_sentence2)

    vowel_counts1, vowel_counts2 = count_vowels(sentence1, sentence2)
    print("\nVowel counts:")
    print("Sentence 1:", ', '.join(f'{vowel}:{count}' for vowel, count in vowel_counts1.items()))
    print("Sentence 2:", ', '.join(f'{vowel}:{count}' for vowel, count in vowel_counts2.items()))

if __name__ == "__main__":
    main()


```

# Question 4 (20 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
import numexpr as ne
import math

def calculate_expression(expression):
    try:
        result = ne.evaluate(expression)
        return result
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
        return None
    except Exception as e:
        print(f"Error: Invalid input expression - {str(e)}")
        return None

def extend_calculator():
    print("Extended Calculator")
    print("1. Exponentiation (x^y)")
    print("2. Square Root (sqrt(x))")
    print("3. Modulus (x % y)")
    choice = input("Enter your choice (1/2/3): ")

    if choice == "1":
        x = float(input("Enter base: "))
        y = float(input("Enter exponent: "))
        result = math.pow(x, y)
    elif choice == "2":
        x = float(input("Enter number: "))
        if x < 0:
            print("Error: Square root of negative number is not allowed.")
            return
        result = math.sqrt(x)
    elif choice == "3":
        x = float(input("Enter dividend: "))
        y = float(input("Enter divisor: "))
        if y == 0:
            print("Error: Division by zero is not allowed.")
            return
        result = x % y
    else:
        print("Error: Invalid choice.")
        return

    print("Result:", result)

def main():
    while True:
        print("\nCommand-Line Calculator")
        print("1. Evaluate Expression")
        print("2. Extended Calculator")
        print("3. Exit")
        choice = input("Enter your choice (1/2/3): ")

        if choice == "1":
            expression = input("Enter mathematical expression: ")
            result = calculate_expression(expression)
            if result is not None:
                print("Result:", result)
        elif choice == "2":
            extend_calculator()
        elif choice == "3":
            print("Exiting calculator. Goodbye!")
            break
        else:
            print("Error: Invalid choice.")

if __name__ == "__main__":
    main()
```

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

To answer this question, you will need to download the file [Assignment1_denton_housing.csv](https://raw.githubusercontent.com/unt-iialab/info5731-spring2022/main/assignments/Assignment1_denton_housing.csv) included with this assignment. The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (10 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (10 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
!pip install pandas
import pandas as pd
from scipy.stats import hmean

# Load the data from CSV
file_name = '/content/sample_data/Assignment1_denton_housing.csv'
df = pd.read_csv(file_name)

# (a) Calculate the difference in the number of Occupied Housing Units from year to year
df_occupied = df[df['title_field'] == 'Occupied Housing Units']
df_occupied_sorted = df_occupied.sort_values(by='year')

yearly_diff = df_occupied_sorted['value'].diff().dropna()
yearly_diff_sorted = yearly_diff.sort_values()

print("\n(a) Difference in Occupied Housing Units (sorted in ascending order):")
for i in range(len(yearly_diff_sorted)):
    print(f"{df_occupied_sorted.iloc[i+1]['year']}: {yearly_diff_sorted.iloc[i]}")

# (b) Calculate the percentage of housing units that are vacant and occupied for each year
print("\n(b) Percentage of vacant and occupied housing units:")

for year in df['year'].unique():
    total_units = df[df['year'] == year]['value'].sum()
    vacant_units = df[(df['year'] == year) & (df['title_field'] == 'Vacant Housing Units')]['value'].values[0]
    occupied_units = df[(df['year'] == year) & (df['title_field'] == 'Occupied Housing Units')]['value'].values[0]

    vacant_percent = (vacant_units / total_units) * 100
    occupied_percent = (occupied_units / total_units) * 100

    print(f"{year} {vacant_percent:.2f}% {occupied_percent:.2f}%")

# (c) Find the year and value for the highest number of vacant and occupied housing units
max_vacant = df[df['title_field'] == 'Vacant Housing Units'].sort_values(by='value', ascending=False).iloc[0]
max_occupied = df[df['title_field'] == 'Occupied Housing Units'].sort_values(by='value', ascending=False).iloc[0]

print("\n(c) Year and value of highest number of housing units:")
print(f"Vacant Housing Units      {max_vacant['year']}   {max_vacant['value']}")
print(f"Occupied Housing Units    {max_occupied['year']}   {max_occupied['value']}")

# (d) Calculate the harmonic mean of the total housing units for all years
total_units_per_year = df.groupby('year')['value'].sum()
harmonic_mean_value = hmean(total_units_per_year)

print(f"\n(d) Harmonic mean of total housing units: {harmonic_mean_value:.2f}")
```
